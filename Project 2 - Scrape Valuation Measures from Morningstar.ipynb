```python
counter = 0
for ticker in df3['Ticker']:
    div, book, NI, OM, EPS, rev = np.NaN, np.NaN, np.NaN, np.NaN, np.NaN, np.NaN
    counter +=1
    try:
        url = 'http://financials.morningstar.com/finan/financials/getFinancePart.html?&callback=xxx&t={stock}'.format(stock = ticker)
        soup = BeautifulSoup(json.loads(re.findall(r'xxx\((.*)\)', requests.get(url).text)[0])['componentData'], 'lxml')
        for i, tr in enumerate(soup.select('tr')):
            row_data = [td.text for td in tr.select('td, th') if td.text]
            if len(row_data) < 12:
                row_data = ['X'] + row_data
            if not row_data: continue
            if row_data[0] == 'Dividends':
                div = row_data[8]
            if row_data[0] == 'Book Value Per Share *\xa0USD':
                book = row_data[8]

            if row_data[0] == 'Net Income\xa0USD Mil':
                NI = row_data[8]

            if row_data[0] == 'Operating Margin %':
                OM = row_data[8]

            if row_data[0] == 'Earnings Per Share\xa0USD':
                EPS = row_data[8]

            if row_data[0] == 'Revenue\xa0USD Mil':
                rev = row_data[8]

        print(counter, ticker, div, book, NI, OM, EPS, rev)

        metrics_df2 = metrics_df2.append({'Ticker': ticker, 'Dividends': div, 'Book_Value': book, 'Net_Income':NI, 'Operating_Margin': OM, 'EPS2': EPS, 'Revenue': rev}, ignore_index=True)

    except:
        print(counter, ticker, div, book, NI, OM, EPS, rev)
        metrics_df2 = metrics_df2.append({'Ticker': ticker, 'Dividends': div, 'Book_Value': book, 'Net_Income':NI, 'Operating_Margin': OM, 'EPS2': EPS, 'Revenue': rev}, ignore_index=True)

```

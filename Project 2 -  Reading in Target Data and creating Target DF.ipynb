### 1. Read in PDF file containing Russell 3000 Data and Extract Company Name and Ticker into Data Frame


```python
import numpy as np
#Following is the code to extract simple Text from pdf using PyPDF2
import pandas as pd
import tabula
# you can find find the pdf file with complete code in below
# read_pdf will save the pdf table into Pandas Dataframe
russell_df = pd.DataFrame(columns = ['Company', 'Ticker', 'Company.1', 'Ticker.1', 'Unnamed: 0',
       'Unnamed: 1'])
for i in np.arange(1,33):
    df = tabula.read_pdf("russell3000_membership_list.pdf", pages = int(i))
    russell_df = russell_df.append(df[0], ignore_index = True)
# in order to print first 5 lines of Table
```

```python
russell_df
```

```python
companies_list_1 = pd.DataFrame(data = zip(russell_df['Company'], russell_df['Ticker']), columns = ['Company', 'Ticker'])
companies_list_2 = pd.DataFrame(data = zip(russell_df['Ticker.1'], russell_df['Unnamed: 0']),columns = ['Company', 'Ticker'])

russell3000_list = (pd.concat([companies_list_1, companies_list_2], ignore_index = True)).dropna()
```

```python
#pickle for faster loading
russell3000_list.to_pickle('russell3000_list.pkl')
```

# Calculate Percentage Difference

```python
russell3000_list = pd.read_pickle('russell3000_list_v3.pkl')
```

```python
def convert_to_float(row):
    try: 
        return float(row)
    except:
        return row
russell3000_list['Open_Price_2020'] = russell3000_list['Open_Price_2020'].apply(convert_to_float)
russell3000_list['Open_Price_2018'] = russell3000_list['Open_Price_2018'].apply(convert_to_float)
```

```python
type(russell3000_list['Open_Price_2018'][0])
```

```python
def calculate_percentage(df):#open_price_2016, open_price_2019
    try:
        return ((df['Open_Price_2020']-df['Open_Price_2018'])/df['Open_Price_2018'])*100
    except:
        return np.NaN
    
#russell3000_list['Percent Increase'] = map(calculate_percentage, russell3000_list['Open_Price_2016'], russell3000_list['Open_Price_2019'])

russell3000_list['Percent Increase'] = russell3000_list.apply(calculate_percentage, axis = 1)
```
